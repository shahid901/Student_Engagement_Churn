# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
students = pd.read_csv('Dataset_Week2.csv')
students.head()
```

```python
# 1. Count the number of NaN values in each column
empty_per_column = students.isna().sum()

# 2. Filter columns that have at least one NaN
columns_with_empty = empty_per_column[empty_per_column > 0]

# 3. Total number of NaN values across the entire dataframe
total_empty = students.isna().sum().sum()

print(f"Columns with empty entries:\n{columns_with_empty}")
print(f"Total number of empty entries: {total_empty}")
```

```python
# Statistical information:

students.describe()
```

## Data Visualization

```python
# Gender distribution
plt.figure(figsize=(8,6))
sns.set_style("white")
ax = sns.countplot(x='Gender', data=students, palette='coolwarm', edgecolor='blue')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()/2, p.get_height()), 
                ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Age of Learner distribution
plt.figure(figsize=(8,6))
sns.histplot(students['Age of Learner'], bins=20, kde=True, color='green', edgecolor='black', alpha=0.7)
plt.title('Age of Learner Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Age of Learner', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(False)
plt.show()
```

```python
# Scatter plot of Age vs. Engagement Score
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age of Learner', y='Engagement Score', data=students, hue='Gender', palette='bright', alpha=0.7, edgecolor='black')
plt.title('Age of Learner vs Engagement Score', fontsize=14, fontweight='bold')
plt.xlabel('Age of Learner', fontsize=12)
plt.ylabel('Engagement Score', fontsize=12)
plt.legend(title='Gender', loc='upper left', fontsize=10, bbox_to_anchor=(0, 1), )
plt.show()
```

```python
# Identifiy increases or decreases in signups using visualisation

# Group data by 'SignUpYear' and 'SignUpMonth' and count signups
signups_by_month = students.groupby(['SignUpYear', 'SignUpMonth'])['First Name'].count().reset_index()

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='SignUpMonth', y='First Name', hue='SignUpYear', data=signups_by_month, marker='o')
plt.title('Monthly Signups Trend')
plt.xlabel('Month')
plt.ylabel('Number of Signups')
plt.xticks(range(1, 13))  # Set x-axis ticks to represent months
plt.grid(True)
plt.show()
```

```python
# Plotting the number of sign-ups in different days of a week
plt.figure(figsize=(9,5))
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(x='SignUpDayOfWeek', data=students, order=day_order, palette='dark')
plt.title('SignUp Day of Week')
plt.xlabel('Days of Week')
plt.ylabel('Number of Students')
plt.show()
```

```python
signup_heatmap = students.groupby(['SignUpYear', 'SignUpMonth'])['Opportunity Id'].count().unstack()
sns.heatmap(signup_heatmap, cmap='Reds', annot=True, fmt='g')
plt.title('Signup Seasonality Heatmap')
plt.xlabel('Signup Month')
plt.ylabel('Signup Year')
plt.show()
```

```python
# Line graph tracking completion rates over time.

# Group data by 'SignUpYear' and 'Opportunity Category' and count completions
# (Replace 'Completion Status' with the actual column representing completion)
completion_trends = students.groupby(['SignUpYear', 'Opportunity Category'])['Opportunity Name'].count().reset_index()

# Rename columns for clarity
completion_trends.columns = ['SignUpYear', 'Opportunity Category', 'Completion Count']

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='SignUpYear', y='Completion Count', hue='Opportunity Category', data=completion_trends, marker='o')
plt.title('Completion Trends over Time by Opportunity Category')
plt.xlabel('Sign Up Year')
plt.ylabel('Number of Completions')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=students['Gender'], y=students['Engagement Score'], hue=students['Gender'],palette="Set3")
plt.title("Gender vs Engagement Score")
plt.xlabel("Gender")
plt.ylabel("Engagement Score")
plt.show()
```

```python
# Box plots illustrating time differences in completion.

# Calculate time differences for completion (replace with your actual completion criteria)
students['Time_to_Completion'] = (pd.to_datetime(students['Opportunity End Date']) - pd.to_datetime(students['Opportunity Start Date'])).dt.days

# Create box plots to illustrate time differences by 'Opportunity Category'
plt.figure(figsize=(12, 6))
sns.boxplot(x='Opportunity Category', y='Time_to_Completion', data=students)
plt.title('Time to Completion by Opportunity Category')
plt.xlabel('Opportunity Category')
plt.ylabel('Time to Completion (Days)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Create a comparison dataframe for signups and completions
signup_completion_comparison = completion_trends.groupby('SignUpYear').agg(
	Signups=('Completion Count', 'sum'),
	Completions=('Completion Count', 'sum')
).reset_index()

# Bar chart comparing signup and completion counts
plt.figure(figsize=(10, 6))
X_axis = np.arange(len(signup_completion_comparison['SignUpYear']))
plt.bar(X_axis - 0.2, signup_completion_comparison['Signups'], 0.4, label='Signups')
plt.bar(X_axis + 0.2, signup_completion_comparison['Completions'], 0.4, label='Completions')

plt.xticks(X_axis, signup_completion_comparison['SignUpYear'])
plt.xlabel("Year")
plt.ylabel("Number of users")
plt.title("Signups vs Completions")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=students, x='Days Since Last Engagement', hue='High Engagement', multiple='stack', bins=20)
plt.title('Days Since Last Engagement for High vs. Low Engagement')
plt.xlabel('Days Since Last Engagement')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Opportunity Category', hue='Status Description', data=students, palette='Set1')
plt.title('Application Status by Opportunity Category')
plt.xlabel('Opportunity Category')
plt.ylabel('Application Status Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
top_countries = students['Country'].value_counts().nlargest(10)
sns.barplot(x=top_countries.index, y=top_countries.values, palette='colorblind')
plt.title('Top 10 Countries by Learner Count')
plt.xlabel('Country')
plt.ylabel('Number of Learners')
plt.xticks(rotation=90)
plt.show()
```

```python
#Correlation between engagement scores and Numerical variables variables.

# Correlation between engagement scores and numerical variables
numerical_cols = students.select_dtypes(include=np.number).columns
correlation_matrix = students[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Calculate and print correlations with 'Engagement Score'
engagement_correlations = correlation_matrix['Engagement Score'].drop('Engagement Score')
print("Correlation between Engagement Score and other numerical features:\n", engagement_correlations)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)  # Assume 3 clusters
students_clustered = students[['Engagement Score', 'Time in Opportunity', 'Days Since Last Engagement']].dropna()
kmeans.fit(students_clustered)

plt.figure(figsize=(8,6))
plt.scatter(students_clustered['Engagement Score'], students_clustered['Time in Opportunity'], 
            c=kmeans.labels_, cmap='viridis')
centers = kmeans.cluster_centers_
plt.title('K-Means Clustering of Students')
plt.xlabel('Engagement Score')
plt.ylabel('Time in Opportunity')
plt.show()
```

```python
# Create the pair plot with customizations
sns.pairplot(
    students[['Time in Opportunity', 'Engagement Score', 'Days Since Last Engagement', 'High Engagement']],
    hue='High Engagement',   # Color points based on 'High Engagement' column
    palette='coolwarm',      # Use a visually distinct color palette
    markers=["o", "s"],      # Different markers for each category in 'High Engagement'
    plot_kws={'alpha': 0.6}, # Set transparency for the points to make overlapping clearer
)

# Add titles and labels
plt.suptitle('Pair Plot of Engagement Features by Engagement Level', fontsize=16)
plt.subplots_adjust(top=0.95)  # Adjust space for the title

# Display the plot
plt.show()
```
