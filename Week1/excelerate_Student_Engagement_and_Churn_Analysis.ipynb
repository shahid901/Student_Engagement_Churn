```python
# Check for missing values in each column
missing_values = SLU.isnull().sum()

# Print the number of missing values for each column
print(missing_values)
```

```python
# Display rows with missing values in 'Institution Name' and 'Current/Intended Major'
missing_institution_major = SLU[SLU['Institution Name'].isnull() | SLU['Current/Intended Major'].isnull()]
print(missing_institution_major[['Institution Name', 'Current/Intended Major']])
```

```python
# Fill missing values in 'Institution Name' with the mode
institution_mode = SLU['Institution Name'].mode()[0]
SLU['Institution Name'].fillna(institution_mode, inplace=True)

# Fill missing values in 'Current/Intended Major' with the mode
major_mode = SLU['Current/Intended Major'].mode()[0]
SLU['Current/Intended Major'].fillna(major_mode, inplace=True)

# Verify that missing values have been filled
missing_values_after_fill = SLU.isnull().sum()
print(missing_values_after_fill)

# Display rows that previously had missing values to confirm the fill
print(SLU.loc[missing_institution_major.index, ['Institution Name', 'Current/Intended Major']])
```

```python
# rows where First Name and Insitution Name contain characters other than alphabets

import re

def contains_non_alpha(text):
  return bool(re.search(r'[^a-zA-Z]', text))

# Apply the function to 'First Name' and 'Institution Name' columns
SLU['NonAlphaFirstName'] = SLU['First Name'].apply(contains_non_alpha)
SLU['NonAlphaInstitution'] = SLU['Institution Name'].apply(contains_non_alpha)

# Filter rows where either column contains non-alphabetic characters
filtered_SLU = SLU[SLU['NonAlphaFirstName'] | SLU['NonAlphaInstitution']]

# Display the filtered rows
filtered_SLU
```

```python
# standardize the  First Name and Insitution Name contain characters other than alphabets

def standardize_text(text):
  # Remove non-alphabetic characters and convert to lowercase
  text = re.sub(r'[^a-zA-Z]', '', text).lower()
  return text

# Apply the function to 'First Name' and 'Institution Name' columns
SLU['First Name'] = SLU['First Name'].apply(standardize_text)
SLU['Institution Name'] = SLU['Institution Name'].apply(standardize_text)
```

```python
# Forward fill missing values in 'Opportunity Start Date'
SLU['Opportunity Start Date'] = SLU['Opportunity Start Date'].ffill()
```

```python
# Check for missing values in each column after the previous operations.
missing_values_final = SLU.isnull().sum()
missing_values_final
```

```python
# Convert specified text columns to title case and strip whitespace
text_columns = ['First Name', 'Country', 'Institution Name', 'Current/Intended Major', 'Opportunity Name', 'Opportunity Category']

#Strip any leading or trailing spaces in column names before accessing
SLU.columns = SLU.columns.str.strip()

for col in text_columns:
    SLU[col] = SLU[col].astype(str).str.title().str.strip()
```

```python
# Standardizing Date Formats
date_cols = ['Learner SignUp DateTime', 'Date of Birth', 'Apply Date', 'Opportunity Start Date', 'Opportunity End Date',]
for col in date_cols:
    SLU[col] = pd.to_datetime(SLU[col], errors='coerce')
```

```python
# Check for missing values (NaT) in the date columns after conversion
for column in date_cols:
    print(f"Number of missing values in {column}:", SLU[column].isna().sum())
```

```python
# Forward fill missing values in specified date columns
for col in ['Learner SignUp DateTime', 'Opportunity Start Date', 'Opportunity End Date']:
    SLU[col] = SLU[col].ffill()
```

## **Feature Egineering**

```python
# Calculate the age of the learner
SLU['Age of Learner'] = (pd.to_datetime('today') - SLU['Date of Birth']).dt.days // 365

# Display the first few rows to verify the age calculation
print(SLU[['Date of Birth', 'Age of Learner']].head())
```

```python
# Calculate engagement duration
SLU['Engagement Duration'] = (SLU['Apply Date'] - SLU['Opportunity Start Date']).dt.days

# Display the first few rows to verify the engagement duration calculation
print(SLU[['Apply Date', 'Opportunity Start Date', 'Engagement Duration']].head())
```

```python
# Count missing values in 'Engagement Duration'
missing_engagement_duration = SLU['Engagement Duration'].isnull().sum()
print(f"Number of missing values in Engagement Duration: {missing_engagement_duration}")
```

```python
# Fill missing values in 'Engagement Duration' with the mean
mean_engagement_duration = SLU['Engagement Duration'].mean()
SLU['Engagement Duration'].fillna(mean_engagement_duration, inplace=True)

# Verify that missing values have been filled
missing_engagement_duration_after_fill = SLU['Engagement Duration'].isnull().sum()
print(f"Number of missing values in Engagement Duration after fill: {missing_engagement_duration_after_fill}")
```

```python
# Extract month and year from 'Learner SignUp DateTime'
SLU['SignUpMonth'] = SLU['Learner SignUp DateTime'].dt.month
SLU['SignUpYear'] = SLU['Learner SignUp DateTime'].dt.year

# Preview the new columns
print(SLU[['Learner SignUp DateTime', 'SignUpMonth', 'SignUpYear']].head())
```

```python
# Calculate the duration of each opportunity
SLU['Time in Opportunity'] = (SLU['Opportunity End Date'] - SLU['Opportunity Start Date']).dt.days

# Display the first few rows to verify the calculation
print(SLU[['Opportunity Start Date', 'Opportunity End Date', 'Time in Opportunity']].head())
```

```python
# calculate an Engagement Score by combining features like Time in Opportunity, Age, and Opportunity Category
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Encode categorical column
encoder = LabelEncoder()
SLU['Opportunity Category Encoded'] = encoder.fit_transform(SLU['Opportunity Category'])

# Normalize using MinMaxScaler
scaler = MinMaxScaler()
SLU[['Time in Opportunity', 'Age of Learner', 'Opportunity Category Encoded']] = scaler.fit_transform(
    SLU[['Time in Opportunity', 'Age of Learner', 'Opportunity Category Encoded']]
)
# Assign weights (adjust based on importance)
w1, w2, w3 = 0.5, 0.3, 0.2  # Example: More weight to Time in Opportunity

# Compute Engagement Score
SLU['Engagement Score'] = (w1 * SLU['Time in Opportunity'] +
                           w2 * SLU['Age of Learner'] +
                           w3 * SLU['Opportunity Category Encoded'])
# Display the results
SLU[['Time in Opportunity', 'Age of Learner', 'Opportunity Category Encoded', 'Engagement Score']].head()
```

```python
# Extract the day of the week from 'Learner SignUp DateTime'
SLU['SignUpDayOfWeek'] = SLU['Learner SignUp DateTime'].dt.day_name()

# Display the first few rows to verify the extraction
print(SLU[['Learner SignUp DateTime', 'SignUpDayOfWeek']].head())
```
